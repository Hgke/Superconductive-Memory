```python
import numpy as np
from matplotlib import pyplot as plt
import sys
sys.path.insert(0, r'C:\Users\Admin\Documents\before_github\SymSC_Python_Norm2')
from Objects import JJ, Sine_current, Pulses, L,Ib, Square_pulse
from FunctionCompiler import *
```

```python
L1 = L.L(loc=[1,2],val=1.5)
L2 = L.L(loc=[2,3],val=1.5)
L3 = L.L(loc=[5,6],val=1.5)
L4 = L.L(loc=[6,7],val=1.5)

Ls1 = L.L(loc=[3,4],val=19.083)
Ls2 = L.L(loc=[4,5],val=19.083)

Ib1 = Ib.Ib(loc=[2],val=0.76)
Ib2 = Ib.Ib(loc=[4],val=0.04)
Ib3 = Ib.Ib(loc=[6],val=0.76)

JJ1 = JJ.JJ(loc=[1,0],A=1.408, c=1)
JJ2 = JJ.JJ(loc=[3,0],A=0.696, c=1)
JJ3 = JJ.JJ(loc=[5,0],A=0.696, c=1)
JJ4 = JJ.JJ(loc=[7,0],A=1.408, c=1)

Pulse_increment_1 = Pulses.Pulses(loc=[1],A=2,t0=100, T = 10000)
Pulse_increment_2 = Pulses.Pulses(loc=[1],A=2,t0=200, T = 10000)
Pulse_increment_3 = Pulses.Pulses(loc=[1],A=2,t0=300, T = 10000)
Pulse_increment_4 = Pulses.Pulses(loc=[1],A=2,t0=400, T = 10000)

Pulse_decrement_1 = Pulses.Pulses(loc=[7],A=2,t0=1000,T = 10000)
Pulse_decrement_2 = Pulses.Pulses(loc=[7],A=2,t0=1200,T = 10000)
Pulse_decrement_3 = Pulses.Pulses(loc=[7],A=2,t0=1400, T = 10000)
Pulse_decrement_4 = Pulses.Pulses(loc=[7],A=2,t0=1600, T = 10000)
Pulse_decrement_5 = Pulses.Pulses(loc=[7],A=2,t0=1800, T = 10000)
Pulse_decrement_6 = Pulses.Pulses(loc=[7],A=2,t0=2000, T = 10000)

Increment = np.array([Pulse_increment_1,Pulse_increment_2,Pulse_increment_3,Pulse_increment_4])
Decrement = np.array([Pulse_decrement_1, Pulse_decrement_2, Pulse_decrement_3, Pulse_decrement_4, Pulse_decrement_5, Pulse_decrement_6])

t = np.arange(0, 2500, 0.1)
fc = FunctionCompiler([L1, L2, L3, L4, Ls1, Ls2, Ib1, Ib2, Ib3, JJ1, JJ2, JJ3, JJ4, *Decrement, *Increment],t)

sol = fc.solve()
```

```python
fig, ax = plt.subplots(dpi=80)
ax.plot(t,JJ1.get_data("V",t,sol))
ax.plot(t,JJ4.get_data("V",t,sol))
ax.legend(['Increment: V_JJ1','Decrement: V_JJ4'])

fig, ax = plt.subplots(dpi=80)
ax.plot(t,JJ2.get_data("P",t,sol))
ax.plot(t,JJ3.get_data("P",t,sol))
ax.legend(['SFQ+','SFQ-'])
```
