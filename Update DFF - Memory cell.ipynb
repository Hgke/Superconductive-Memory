```python
import numpy as np
from matplotlib import pyplot as plt
import sys
sys.path.insert(0, r'C:\Users\Admin\Documents\before_github\SymSC_Python_Norm2')
from Objects import JJ, Sine_current, Pulses, L,Ib, Square_pulse
from FunctionCompiler import *
```

```python
L1 = L.L(loc=[1,2],val=1)
L2 = L.L(loc=[6,7],val=1)

Ls1 = L.L(loc=[3,4],val=3)
Ls2 = L.L(loc=[4,5],val=3)

Ib1 = Ib.Ib(loc=[4],val=0.04)

JJ1 = JJ.JJ(loc=[2,3],A=1.96, c=1)
JJ2 = JJ.JJ(loc=[3,0],A=0.9, c=1)
JJ3 = JJ.JJ(loc=[5,0],A=0.9, c=1)
JJ4 = JJ.JJ(loc=[5,6],A=1.96, c=1)

Pulse_increment_1 = Pulses.Pulses(loc=[1],A=2,t0=100, T = 10000)
Pulse_increment_2 = Pulses.Pulses(loc=[1],A=2,t0=300, T = 10000)
Pulse_increment_3 = Pulses.Pulses(loc=[1],A=2,t0=500, T = 10000)
Pulse_increment_4 = Pulses.Pulses(loc=[1],A=2,t0=700, T = 10000)

Pulse_decrement_1 = Pulses.Pulses(loc=[7],A=2,t0=1000,T = 10000)
Pulse_decrement_2 = Pulses.Pulses(loc=[7],A=2,t0=1200,T = 10000)
Pulse_decrement_3 = Pulses.Pulses(loc=[7],A=2,t0=1400, T = 10000)
Pulse_decrement_4 = Pulses.Pulses(loc=[7],A=2,t0=1600, T = 10000)
Pulse_decrement_5 = Pulses.Pulses(loc=[7],A=2,t0=1800, T = 10000)
Pulse_decrement_6 = Pulses.Pulses(loc=[7],A=2,t0=2000, T = 10000)

Increment = np.array([Pulse_increment_1,Pulse_increment_2])
Decrement = np.array([Pulse_decrement_1,Pulse_decrement_2])

t = np.arange(0, 2500, 0.1)
fc = FunctionCompiler([L1, L2, Ls1, Ls2, Ib1, JJ1, JJ2, JJ3, JJ4, *Decrement, *Increment],t)

sol = fc.solve()

fig, ax = plt.subplots(nrows=4,ncols=1,dpi=300,gridspec_kw={'height_ratios': [40,40,40,40]})
#ax.yaxis.set_major_locator(ticker.MultipleLocator(1))
# fig = plt.figure(constrained_layout=False)
# gs = fig.add_gridspec(nrows=6, ncols=1, left=0.05, right=0.48, wspace=0.05)

plt.subplot(4,1,1)
plt.ylabel('Input_1',rotation=0,fontsize=10,labelpad=25)
plt.xticks([])
plt.yticks([0,1,2])
# plt.tick_params(axis = 'both',    #  Применяем параметры к обеим осям
#                pad = -1,    #  Расстояние между черточкой и ее подписью
#                labelsize = 9)
plt.plot(t,Pulse_increment_1.get_data("I",t,sol)+Pulse_increment_2.get_data("I",t,sol))

plt.subplot(4,1,2)
plt.yticks([0,1,2])
plt.ylabel('Input_2',rotation=0,fontsize=10,labelpad=25)
plt.xticks([])

plt.plot(t,Pulse_decrement_1.get_data("I",t,sol)+Pulse_decrement_2.get_data("I",t,sol))

plt.subplot(4,1,3)
plt.ylabel('$P_{J2}$',rotation=0,fontsize=10,labelpad=15)
plt.yticks([0,np.round(2*np.pi,2),np.round(4*np.pi,2)])
plt.xticks([])

plt.plot(t,JJ2.get_data("P",t,sol))

plt.subplot(4,1,4)
plt.ylabel('$P_{J3}$',rotation=0,fontsize=10,labelpad=15)
plt.yticks([0,np.round(2*np.pi,2),np.round(4*np.pi,2)])
plt.plot(t,JJ3.get_data("P",t,sol))
```
